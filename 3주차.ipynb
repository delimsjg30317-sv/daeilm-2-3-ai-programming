반복문과 조건문 함께 사용

```python
scores = [42, 60, 77, 89, 20]
for score in scores:
    if score >= 60:
        print(f'{score}점은 합격입니다.')
    else:
        print(f'{score}점은 불합격입니다.')
    


# 60점은 합격입니다.
# 42점은 불합격입니다.
```

```python
numbers = [1,2,3,4,5]

total = 0
for number in numbers:
    total += number
print("합계:", total)

# 합계: x
```

```python
scores = [65, 82, 91, 100, 58, 77]

# 70점 이상은 합격
# 출력: 합격자는 x명 입니다.

count = 0

for a in scores:
    if a >= 70:
        count += 1

print(f'합격자는 {count} 명입니다.')
```
