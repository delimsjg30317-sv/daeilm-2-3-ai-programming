```python
name = "    Python "

print(f"원본: '{name}'") # JAVA: name.trim()
print(f"정리 후: '{name.strip()}")

# 변수와 String 비교
print("공백이 있을 때 비교:", name == "Python")
print(" 공백 제거 후 비교:", name.strip() == "Python")
```

```python
# 할인율 계산 
# discount: 0.0~1.0 (0%~100%)
def calculate_price(price, discount=0):
    discounted = price * (1-discount)
    return price

print(f"기본 가격: {calculate_price(10000)}원")
print(f"할인 가격: {calculate_price(10000, 0.1)}원")
```

# 함수에서 여러 값 반환하기

```python
# 최솟값과 최댓값을 동시 반환하는 함수
def min_max(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    return minimum, maximum

scores = [10,20,30,40,50,60,70,80,]
low, high = min_max(scores)    
print("최솟값:", low)
print("최대값:", high)
```

# Pandas 심화

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
# 학생 데이터
grade_data = {
    "name": ["kim", "lee", "park", "choi", "jung", "han"],
    "korean": [88, 72, 95, 65, 80, 91   ],
    "math": [75, 90, 65, 85, 77, 82   ],
    "english": [82, 88, 76, 70, 95, 78   ],

}
grade_df = pd.DataFrame(grade_data)
grade_df
```

## 정렬하기

```python
print ("---- 국어 점수 낮은 순 ----")
print(grade_df.sort_values("korean"))
print()
print("----수학 점수 높은 순----")
grade_df.sort_values("math", ascending=False)
```

## 새로운 열 추가하기

```python
# \ 내려쓰기
grade_df["average"] = \
      (grade_df["korean"] + \
       grade_df["math"] + \
       grade_df["english"]) / 3
grade_df["average"] = grade_df["average"].round(1)
grade_df
```

```python
grade_df["grade"] = grade_df["average"].apply(

    lambda x: "A" if x >= 80 else "B" 
)
grade_df
```

## 그룹별 통계

```python
group_avg = grade_df.groupby("grade")["average"].mean()
print("등급별 평균:")
print(group_avg)
```

```python
# 응용 - 회사 부서별 급여 파악

employee_data = {
"name": ["kim", "lee", "park", "choi", "jung", "han"],
"department": ["개발", "마케팅", "개발", "인사", "마케팅", "인사"],
"salary": [4200, 3500, 4800, 3200, 3800, 5000]

}

emp_df = pd.DataFrame(employee_data)

#부서별 평균 급여
dept_avg = emp_df.groupby("department")["salary"].mean()
print("---- 부서별 평균 급여 ----")
print(dept_avg.round(0))
print()
print("---- 부서별 인원 수 ----")
dept_count = emp_df.groupby("department")["name"].count()
print(dept_count)
```
